<h2>归一化</h2>

<h3>train文件</h3>

```python
columns_train = train.columns.tolist()[20:-1]

# 循环遍历每一列，分别进行归一化处理
for column_name in columns_train:
    column_data = train[column_name]
    min_value = column_data.min()
    max_value = column_data.max()
    normalized_data = (column_data - min_value) / (max_value - min_value)
    train[column_name] = normalized_data

# 打印处理后的 DataFrame
train
```

<h3>test文件</h3>

```python
columns_test = test.columns.tolist()[20:-1]

# 循环遍历每一列，分别进行归一化处理
for column_name in columns_test:
    column_data = test[column_name]
    min_value = column_data.min()
    max_value = column_data.max()
    normalized_data = (column_data - min_value) / (max_value - min_value)
    test[column_name] = normalized_data

# 打印处理后的 DataFrame
test
```
