<h2>小提琴图</h2>

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import io
import pyperclip
from PIL import Image
import base64

# 读取数据集
df = pd.read_excel(r'D:\code\数学建模模版\数据集\train33.xlsx')

# 设置图表样式
sns.set(style="whitegrid")

# 绘制小提琴图
plt.figure(figsize=(10, 6))
sns.violinplot(x="target", y='HM_volume', data=df)

# 设置图表标题和轴标签
plt.title('Violin Plot')
plt.xlabel('90mrs')
plt.ylabel('HM_Volumn_Value')

# 将图保存为文件对象
buffer = io.BytesIO()
plt.savefig(buffer, format='png')
buffer.seek(0)

# 将图像转换为 base64 编码的字符串
image_data = base64.b64encode(buffer.getvalue()).decode('utf-8')

# 将图像字符串复制到剪贴板
pyperclip.copy(image_data)

# 显示图表
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import io
import pyperclip
from PIL import Image
import base64

# 读取数据集
df = pd.read_excel(r'D:\code\数学建模模版\数据集\train33.xlsx')

# 创建空的 DataFrame 对象
pd_temp = pd.DataFrame()
pd_temp['ED_volume'] = df['ED_volume']
pd_temp['HM_volume'] = df['HM_volume']
pd_temp['target'] = df['target']

# 获取列名列表
column = pd_temp.columns.tolist()[:-1]

# 设置图表样式
sns.set(style="whitegrid")

# 创建绘图对象
fig = plt.figure(figsize=(20, 6), dpi=70)

# 遍历每个列名，绘制小提琴图
j = 1
for col in column:
    plt.subplot(1, 2, j)
    sns.violinplot(x="target", y=col, data=pd_temp)
    plt.ylabel(col, fontsize=10)
    j += 1

# 将图保存为文件对象
buffer = io.BytesIO()
plt.savefig(buffer, format='png')
buffer.seek(0)

# 将图像转换为 base64 编码的字符串
image_data = base64.b64encode(buffer.getvalue()).decode('utf-8')

# 将图像字符串复制到剪贴板
pyperclip.copy(image_data)

# 打印提示信息
print("图像已复制到剪贴板")

# 显示图表
plt.show()
```
