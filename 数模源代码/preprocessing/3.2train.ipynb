<h2>时间率变化</h2>

<h3>HM变化</h3>

```python
biao_time=pd.read_excel(r"C:\Users\y\Desktop\附表1-检索表格-流水号vs时间(1).xlsx").iloc[:100,:]
```

```python
filtered_cols = biao22.filter(regex='HM.*volume').columns
filtered_cols
```

```python
df_rH=pd.DataFrame()
for i in filtered_cols:
    df_rH[i]=biao22[i]
    
```

```python
time_clu=biao_time.filter(regex='时间点').columns
repreat_clu=biao_time['重复次数']
df_time=pd.DataFrame()
for i in time_clu:
    df_time[i]=biao_time[i]
```

```python
list_rH=[]
x=0
for i in repreat_clu:
    sum=0
    if i>8:
        i=8
    for j in range(i-1):
        t=(df_time.iloc[x,j+1]-df_time.iloc[x,j]).total_seconds() / 3600
        v=df_rH.iloc[x,j+1]-df_rH.iloc[x,j]
        b=abs(v*1.0/t)
        sum+=b
    sum=round(sum * 10**-3, 3)
    x+=1
    list_rH.append(sum)
pd_temp=pd.DataFrame(list_rH)
test['HM_r']=pd_temp
test
```

<h3>ED变化率</h3>

```python
filtered_cols = biao22.filter(regex='ED.*volume').columns
df_rE=pd.DataFrame()
for i in filtered_cols:
    df_rE[i]=biao22[i]
list_rE=[]
x=0
for i in repreat_clu:
    sum=0
    if i>8:
        i=8
    for j in range(i-1):
        t=(df_time.iloc[x,j+1]-df_time.iloc[x,j]).total_seconds() / 3600
        v=df_rE.iloc[x,j+1]-df_rE.iloc[x,j]
        b=abs(v*1.0/t)
        sum+=b
    sum=round(sum * 10**-3, 3)
    x+=1
    list_rE.append(sum)
pd_temp=pd.DataFrame(list_rE)
test['ED_r']=pd_temp
test
```

<h1>表三进行操作</h1>

<h2>子表1进行操作</h2>

```python
ls_clu=biao_time.filter(regex='流水号').columns
df_ls =pd.DataFrame()
for i in ls_clu:
    df_ls[i]=biao_time[i]
df_ls=df_ls.fillna(0)
df_ls
```

<h3>获取ED列</h3>

```python
list_sH=[]
biao32=biao31.iloc[:,1:33]
columns_3z1=biao32.columns.tolist()
pca = PCA(n_components=1)

for i in range(4):
    biao_p=df_ls.iloc[:,i]
    j=0
    df = pd.DataFrame(columns=columns_3z1[1:])
    for a in biao_p:
        rox= biao32.loc[biao32['流水号'] == a]
        if rox.empty:
            df.loc[j]=0
        else:
            df.loc[j]=rox.iloc[0,1:].tolist()
        j+=1
    new_train_pca_90 = pca.fit_transform(df)
    list_sH.append(new_train_pca_90)
data_array = np.array(list_sH)

# 将数据数组重新调整为二维数组
data1 = data_array.reshape(4, 100)

#转置
df_3D=pd.DataFrame(data1.T)
new_train_pca_90 = pca.fit_transform(df_3D)
df_4D=pd.DataFrame(new_train_pca_90)
test['ED_S']=df_4D
test   
```
